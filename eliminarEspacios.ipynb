```python

def eliminarEspacios(str_list):
  resultado = []
  for string in str_list:
    resultado.append(string.replace(" ", ""))
  return resultado

lista = []
print("ingrese la cantidad de elementos a ingresar")
palabra = input()
for x in range(int(palabra)):
  palabra = input()
  lista.append(palabra)


print("Sin espacios")

for x in eliminarEspacios(lista):
  print(x)








```
